```python
# https://stackoverflow.com/questions/49320201/sympy-plotting-in-the-jupyter-notebook-with-inline-backend
from IPython.display import display, Math, Latex
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import sympy 
from sympy.plotting import plot
#sympy.init_printing(use_latex=True)
```

```python
func_str = "t**4 + 3*t**3 - 12*t**2 + t - 6"
t = sympy.symbols("t")
f = sympy.sympify(func_str)
sympy.pprint(f)
```

```python
f_tex = sympy.latex(f, mode="inline")
sympy.pprint(f_tex)
Latex(f_tex)
```

```python
ax = plot(f, show=False)
ax[0].line_color = "r"
ax[0].label = f_tex
ax.legend = True
ax.title = f_tex
ax.show()
```

```python
g_str = "sin(x) - cos(x)"
x = sympy.symbols("x")
g = sympy.sympify(g_str)
sympy.pprint(g)
```

```python
g_simp = sympy.trigsimp(g)
sympy.pprint(g_simp)
g_simp_tex = sympy.latex(g_simp, mode="inline")
Latex(g_simp_tex)
```

```python
from sympy import sin, cos
ax = plot(g, sin(x), -cos(x), show=False)
ax[0].line_color = "r"
ax[1].line_color = "g"
ax[2].line_color = "b"
ax[0].label = g_simp_tex
ax[1].label = "sin"
ax[2].label = "-cos"
ax.legend = True
ax.show()
```

```python
from sympy import exp, ln
#th = sympy.symbols("theta")
al = sympy.symbols("alpha")
theta_func =  exp(al) + al * ln(al)
sympy.pprint(theta_func)
```

```python
theta_func_tex = sympy.latex(theta_func, mode="inline")
Latex(theta_func_tex)
```

```python
ax = plot(theta_func, exp(al), al*ln(al), show=False)
#ax = plot(al*ln(al), (al, -100, 10000),show=False)
ax[0].line_color = "r"
ax[1].line_color = "g"
ax[2].line_color = "b"
ax[0].label = theta_func_tex
ax[1].label = sympy.latex( exp(al), mode="inline")
ax[2].label = sympy.latex( al*ln(al), mode="inline")
ax.legend = True
ax.show()
```

```python
ax = plot(al*ln(al), (al, 0, 10), axis_center=(0,0) , show=False)
ax[0].line_color = "r"
ax[0].label = sympy.latex( al*ln(al), mode="inline")
ax.legend = True
ax.show()
```

```python
#fig, ax = plt.subplots()  # does not work
ax = plot((theta_func, (al, 0, 10)), (exp(al), (al, -10, 10)), show=False)
#ax = plot(al*ln(al), (al, -100, 10000),show=False)
ax[0].line_color = "r"
ax[1].line_color = "g"
ax[0].label = theta_func_tex
ax[1].label = sympy.latex( exp(al), mode="inline")
ax.legend = True
ax.show()
```

```python
ax = plot(theta_func, (al, -0, 10), show=False)
#ax = plot(al*ln(al), (al, -100, 10000),show=False)
ax[0].line_color = "r"
ax[0].label = theta_func_tex
ax.legend = True
ax.show()
```
