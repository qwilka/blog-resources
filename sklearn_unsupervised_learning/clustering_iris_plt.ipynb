```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
iris_df = datasets.load_iris()
print("iris_df.target=", iris_df.target)
```

```python
model = KMeans(n_clusters=3, init='k-means++', max_iter=100, n_init=1, verbose=0, random_state=0)
model.fit(iris_df.data)
```

```python
predicted_label = model.predict([[7.2, 3.5, 0.8, 1.6]])
print("predicted_label", predicted_label)
```

```python
all_predictions = model.predict(iris_df.data)
print("all_predictions", all_predictions)
```

```python
x_axis = iris_df.data[:, 0]  # Sepal Length
y_axis = iris_df.data[:, 2]  # Sepal Width
cc = ['red','green','blue']
cmap = ListedColormap(cc)
fig, (ax1, ax2) = plt.subplots(ncols=2)  # sharey=True
ax1.scatter(x_axis, y_axis, c=iris_df.target, cmap=cmap)
# https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html
#Â https://matplotlib.org/gallery/lines_bars_and_markers/scatter_with_legend.html
ax1.legend(labels=['red','green','blue'], scatterpoints=1, title='testing')
#ax1.yaxis.tick_right()
ax2.scatter(x_axis, y_axis, c=all_predictions, cmap=cmap)
```
